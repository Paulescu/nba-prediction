```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 500)

# For Visualization
import matplotlib.pyplot as plt
import seaborn as sns
#%config InlineBackend.figure_format = 'svg'

from pathlib import Path  #for Windows/Linux compatibility
DATAPATH = Path(r'data')
```

```python
df_traditional = pd.read_csv(DATAPATH / "traditional_seasons.csv")
df_traditional.head()
```

```python
df_advanced = pd.read_csv(DATAPATH / "advanced_seasons.csv")
df_advanced.head()
```

```python
df_four_factors = pd.read_csv(DATAPATH / "four-factors_seasons.csv")
df_four_factors.head()
```

```python
df_misc = pd.read_csv(DATAPATH / "misc_seasons.csv")
df_misc.head()
```

```python
df_scoring = pd.read_csv(DATAPATH / "scoring_seasons.csv")
df_scoring.head()
```

```python
def fix_spaces(df):
    df.columns = df.columns.str.replace(' ', '_')
    print(df.columns)
    return df

#df_traditional = fix_spaces(df_traditional)
#df_advanced = fix_spaces(df_advanced)
df_four_factors = fix_spaces(df_four_factors)
df_misc = fix_spaces(df_misc)
df_scoring = fix_spaces(df_scoring)

df = fix_spaces(df_four_factors)
df
```

```python
traditional_columns = ['PTS', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'TOV', 'STL', 'BLK', 'PF', '+/-']

advanced_columns = ['OffRtg', 'DefRtg', 'NetRtg', 'AST%', 'AST/TO', 'ASTRatio', 'OREB%', 'DREB%', 'REB%', 'TOV%', 'eFG%', 'TS%', 'PACE', 'PIE', 'TEAM_ID', 'GAME_ID']	

four_factors_columns = ['FTARate', 'OppeFG%', 'OppFTA_Rate', 'OppTOV%', 'OppOREB%','TEAM_ID', 'GAME_ID'] # duplicates = ['eFG%', 'TOV%', 'OREB%',]

misc_columns = ['PTSOFF_TO', '2ndPTS', 'FBPs', 'PITP', 'Opp_PTSOFF_TO', 'Opp2nd_PTS', 'OppFBPs', 'OppPITP','TEAM_ID', 'GAME_ID']

scoring_columns = ['%FGA2PT', '%FGA3PT', '%PTS2PT', '%PTS2PT_MR', '%PTS3PT', '%PTSFBPs', '%PTSFT', '%PTSOff_TO', '%PTSPITP', '2FGM%AST', '2FGM%UAST', '3FGM%AST', '3FGM%UAST', 'FGM%AST', 'FGM%UAST','TEAM_ID', 'GAME_ID']

```

```python
# combine all dataframes together
df = df_traditional.merge(df_advanced[advanced_columns], how='inner', on=['GAME_ID','TEAM_ID'])
df = df.merge(df_four_factors[four_factors_columns], how='inner', on=['GAME_ID','TEAM_ID'])
df = df.merge(df_misc[misc_columns], how='inner', on=['GAME_ID','TEAM_ID'])
df = df.merge(df_scoring[scoring_columns], how='inner', on=['GAME_ID','TEAM_ID'])
df
```
